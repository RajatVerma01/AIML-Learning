```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Housing.csv")
```

```python
df.head()
```

## Exploratary data analysis

By performing EDA you can deeply understand the insights present in your data to improve the performance of your models. Here you analyze features and their effect on the target. By performing eda you are able to find insights that can help you develop high-performing models.

### Univariate analysis

#### 1. Numerical columns

```python
num=[]
cat=[]
for i in df.columns:
        if df[i].dtype=='object':
            cat.append(i)
        else:
            num.append(i)
```

```python
cat
```

```python
num
```

```python
for col in num:
    print(f"{col} value counts", df[col].value_counts())
for col in cat:
    print(f"{col} value counts", df[col].value_counts())
```

```python
for col in num:
    print(f"{col} unique values", df[col].nunique())
for col in cat:
    print(f"{col} value values", df[col].nunique())
```

```python
for col in num:
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
```

Here we can check the distribution of data that whether the data is normally distributed or it is skewed.

Skewed data needs to be treated accordingly.

```python
for col in num:
    sns.boxplot(x=df[col])
    plt.title(f"Box Plot of {col}")
    plt.show()
```

By drawing box plot you can find out the outliers in the data which can affect the performance of the model. 
These outliers need to be treated accordingly.

#### 2. Categorical columns

```python
for col in cat:
    sns.countplot(x=col, data=df)
    plt.xticks(rotation=45)
    plt.title(f"distribution of {col}")
    plt.show()
```

## Bivariate Analysis

#### 1. Numerical column

```python
for col in num:
    if col != 'price':
        plt.figure(figsize=(4,4))
        sns.scatterplot(x=df[col], y=df['price'])
        plt.title(f'{col} vs price')
        plt.show()
```

By performing bivariate analysis, we can understand the relationship between a feature (independent variable) and the target (dependent variable), which helps in identifying how strongly a feature influences the target.”

#### 2. Categorical data

```python
for col in cat:
    sns.countplot(x=col, hue='price', data=df)
    plt.title(f"{col} vs Price")
    plt.show()
```

## Multivariate analysis

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

By performning multi variate analysis you can find the correlation of features with other features and features with target which helps to understand that which feature will have the highest impact on the target.
