# Here we have learned that how we can gather data by web scraping.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
responseurl = "http://quotes.toscrape.com/"
response = requests.get(responseurl)
soup = BeautifulSoup(response.text, "html.parser")
quotes_data = []

quotes = soup.find_all("div", class_="quote")

for q in quotes:
    text = q.find("span", class_="text").text
    author = q.find("small", class_="author").text
    tags = [tag.text for tag in q.find_all("a", class_="tag")]
    
    quotes_data.append({
        "quote": text,
        "author": author,
        "tags": ", ".join(tags)
    })
```

```python
df=pd.DataFrame(quotes_data)
```
