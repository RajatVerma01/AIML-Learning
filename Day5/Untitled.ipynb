## Simple linear regression 

In Simple Linear Regression, we try to find the best-fit line that represents the relationship between the input variable (X) and the output variable (Y).

The model calculates the values of the coefficients m (slope) and b (intercept) in the linear equation:
m (slope) represents how much Y changes when X increases by one unit.
b (intercept) represents the value of Y when X = 0.
Once the best values of m and b are found, the equation can be used to predict the value of Y for new input values of X.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('simple.csv')
```

```python
df.head()
```

```python
plt.scatter(df['x'],df['y'])
plt.xlabel('x')
plt.ylabel('(y)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
plt.scatter(df['x'],df['y'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('x')
plt.ylabel('y')
```

Here the red line shows the line that fits on the data

```python
m=lr.coef_
m
```

```python
b=lr.intercept_
b
```
