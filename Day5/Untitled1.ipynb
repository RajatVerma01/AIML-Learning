## Multiple Linear regression

In Multiple Linear Regression, we try to find the best-fit relationship between one dependent variable (Y) and multiple independent variables (X₁, X₂, X₃, ...).

Instead of using only one input feature like Simple Linear Regression, Multiple Linear Regression uses several features together to predict the output more accurately.

The equation of Multiple Linear Regression is:
y=m1*x1+m2*x2+......+b

where m1,m2,m3...... are coefecient of features
b is the intercept
and x1,x2,x3....  are the input variables
